```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
```

```python
interactions = pd.read_csv('temp/interactions.csv')
preferences = pd.read_csv('temp/preferences.csv')
```

```python
label_encoder_user = LabelEncoder()
label_encoder_storybook = LabelEncoder()

interactions['user_id'] = label_encoder_user.fit_transform(interactions['user_id'])
interactions['storybook_id'] = label_encoder_storybook.fit_transform(interactions['storybook_id'])
```

```python
interaction_matrix = interactions.pivot_table(index='user_id', columns='storybook_id', values='interaction_type', aggfunc='count').fillna(0)
interaction_sparse_matrix = csr_matrix(interaction_matrix.values)
```

```python
train_data, test_data = train_test_split(interaction_matrix, test_size=0.2, random_state=42)
```

```python
model = NearestNeighbors(metric='cosine', algorithm='brute')
model.fit(train_data)
```

```python
def recommend_storybooks(user_id, n_recommendations=5):
    if not interactions['user_id'].isin([user_id]).any():
        return []

    user_index = label_encoder_user.transform([user_id])[0]
    distances, indices = model.kneighbors([interaction_matrix.iloc[user_index].values], n_neighbors=n_recommendations + 1)

    print(f"User ID: {user_id}, User Index: {user_index}")
    print(f"Distances: {distances}")
    print(f"Indices: {indices}")

    recommendations = []
    for i in range(1, len(distances.flatten())):
        recommended_index = indices.flatten()[i]
        if recommended_index >= len(interaction_matrix.columns):
            print(f"Warning: Index {recommended_index} out of bounds for interaction_matrix.columns")
            continue
        recommendations.append(interaction_matrix.columns[recommended_index])

    recommended_storybooks = list(label_encoder_storybook.inverse_transform(recommendations))
    print(f"Recommended Storybooks: {recommended_storybooks}")

    return recommended_storybooks


# Evaluate the model
def evaluate_model(test_data):
    test_users = test_data.index.tolist()
    all_true_positives = 0
    all_false_positives = 0
    all_false_negatives = 0

    for user_index in test_users:
        try:
            actual_storybooks = set(test_data.columns[test_data.loc[user_index] > 0].tolist())
        except IndexError as e:
            print(f'Error accessing test_data for user_index {user_index}: {e}')
            continue

        user_id = label_encoder_user.inverse_transform([user_index])[0]
        recommended_storybooks = set(recommend_storybooks(user_id, n_recommendations=5))

        print(f'User ID: {user_id}')
        print(f'Actual Storybooks: {actual_storybooks}')
        print(f'Recommended Storybooks: {recommended_storybooks}')

        true_positives = len(actual_storybooks.intersection(recommended_storybooks))
        false_positives = len(recommended_storybooks.difference(actual_storybooks))
        false_negatives = len(actual_storybooks.difference(recommended_storybooks))

        print(f'True Positives: {true_positives}, False Positives: {false_positives}, False Negatives: {false_negatives}')

        all_true_positives += true_positives
        all_false_positives += false_positives
        all_false_negatives += false_negatives

    precision = all_true_positives / (all_true_positives + all_false_positives) if (all_true_positives + all_false_positives) > 0 else 0
    recall = all_true_positives / (all_true_positives + all_false_negatives) if (all_true_positives + all_false_negatives) > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    print(f'Precision: {precision:.2f}')
    print(f'Recall: {recall:.2f}')
    print(f'F1 Score: {f1:.2f}')
```

```python
# Example plotting precision and recall
plt.figure(figsize=(10, 6))
plt.plot([1, 2, 3], [0.5, 0.7, 0.8], marker='o', linestyle='-', color='b', label='Precision')
plt.plot([1, 2, 3], [0.6, 0.8, 0.7], marker='x', linestyle='--', color='r', label='Recall')
plt.xlabel('Iteration')
plt.ylabel('Score')
plt.title('Precision and Recall')
plt.legend()
plt.grid(True)
plt.show()
```

```python
evaluate_model(test_data)
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import OneHotEncoder

# Example datasets (dummy data for illustration)
interactions = pd.DataFrame({
    'user_id': ['user1', 'user1', 'user2', 'user2', 'user3'],
    'item_id': ['item1', 'item2', 'item1', 'item3', 'item2'],
    'rating': [5, 4, 3, 5, 2]
})

item_features = pd.DataFrame({
    'item_id': ['item1', 'item2', 'item3'],
    'genre': ['Fantasy', 'Mystery', 'Science Fiction'],
    'author': ['Author1', 'Author2', 'Author3'],
    'keywords': ['magic, adventure', 'detective, suspense', 'space, aliens']
})

# One-hot encode categorical features
encoder = OneHotEncoder()
encoded_features = encoder.fit_transform(item_features[['genre', 'author', 'keywords']]).toarray()

# Create a DataFrame with the encoded features
encoded_item_features = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out(['genre', 'author', 'keywords']))
encoded_item_features['item_id'] = item_features['item_id']

# Function to extract features for content-based filtering
def extract_features(item_id):
    return encoded_item_features[encoded_item_features['item_id'] == item_id].drop('item_id', axis=1).values.flatten()

# Collaborative filtering function
def collaborative_filtering(user_id, n_recommendations=5):
    user_interactions = interactions[interactions['user_id'] == user_id]
    user_items = user_interactions['item_id'].values
    
    # Example: Recommend items liked by similar users
    similar_users = interactions[interactions['item_id'].isin(user_items)]['user_id'].unique()
    
    # Select items liked by similar users that the target user hasn't interacted with
    recommendations = interactions[~interactions['user_id'].isin([user_id])]\
                         [interactions['user_id'].isin(similar_users)]\
                         .groupby('item_id').size().sort_values(ascending=False).head(n_recommendations).index
    
    return recommendations.tolist()

# Content-based filtering function
def content_based_filtering(user_id, n_recommendations=5):
    # Get items interacted with by the user
    user_interactions = interactions[interactions['user_id'] == user_id]
    user_items = user_interactions['item_id'].values
    
    # Extract features for user's interacted items
    user_preferences = np.array([extract_features(item_id) for item_id in user_items])
    
    # Extract features for all items
    item_ids = encoded_item_features['item_id'].values
    item_features_matrix = np.array([extract_features(item_id) for item_id in item_ids])
    
    # Calculate cosine similarity between user preferences and item features
    similarities = cosine_similarity(user_preferences, item_features_matrix)
    
    # Find items most similar to user preferences
    top_indices = similarities.mean(axis=0).argsort()[::-1][:n_recommendations]
    recommended_items = item_ids[top_indices]
    
    return recommended_items.tolist()

# Hybrid recommendation function
def hybrid_recommendation(user_id, n_recommendations=5):
    cf_recommendations = collaborative_filtering(user_id, n_recommendations)
    cb_recommendations = content_based_filtering(user_id, n_recommendations)
    
    # Combine recommendations (simple union for illustration)
    hybrid_recommendations = list(set(cf_recommendations).union(set(cb_recommendations)))
    
    return hybrid_recommendations

# Example usage:
user_id = 'user1'
recommended_books = hybrid_recommendation(user_id, n_recommendations=5)
print(f'Recommended books for {user_id}: {recommended_books}')
```
